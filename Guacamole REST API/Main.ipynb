### Initialize OAuth

```python
import gspread
from oauth2client.service_account import ServiceAccountCredentials

# Define the scope and credentials
scope = ['https://spreadsheets.google.com/feeds',
         'https://www.googleapis.com/auth/drive']
creds = ServiceAccountCredentials.from_json_keyfile_name('service_account_secret.json', scope)

# Authorize the client
client = gspread.authorize(creds)

# Open the Google Sheet by its ID
sheet = client.open_by_key('127tfOUFWpw5MMWCanvuNBexCkUhFnshiWSMvn3F3N1c')
```

### Parse spreadsheet data into dictionaries with login credentials

```python
# Select the "Logins" spreadsheet
worksheet = sheet.worksheet('Logins')

# Get all values from columns A/B and C/D
usernames1 = worksheet.col_values(1)
passwords1 = worksheet.col_values(2)
usernames2 = worksheet.col_values(3)
passwords2 = worksheet.col_values(4)

# Create key-value pairs
login_info1 = dict(zip(usernames1[:], passwords1[:]))
login_info2 = dict(zip(usernames2[:], passwords2[:]))
```

### Connect to the API and getting the authentication token for the admin user

```python
import requests
url = "http://localhost:8080/api/"
headers = {
    'Accept': 'application/json',
    'Content-Type': 'application/x-www-form-urlencoded'
}
res = requests.post(url + "tokens", headers=headers, data={'username': 'anish', 'password': 'Madelia1$'})
masterToken = res.json()['authToken']
```

### Create Team 1 Accounts

```python
headers = {
          'Accept': 'application/json',
          'Content-Type': 'application/json;charset=UTF-8'
}

for username, password in login_info1.items():
    uri = "session/data/postgresql/users?token=" + masterToken
    data = {
            "username": username,
            "password": password,
            "attributes": {"disabled":"","expired":"","access-window-start":"","access-window-end":"","valid-from":"","valid-until":"","timezone":None}
    }
    res = requests.post(url + uri, headers=headers, json=data)
    
    uri = "session/data/postgresql/users/" + username + "/permissions?token=" + masterToken
    data = [
              {
                "op": "add",
                "path": "/systemPermissions",
                "value": "CREATE_CONNECTION"
              }
    ]
    res = requests.patch(url + uri, headers=headers, json=data)
```
